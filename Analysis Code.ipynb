# Q18)Are there any correlations between different attributes and the likelihood of subscribing to a term deposit?

```python
df['y']=df['y'].map({'yes':1,'no':0})
```

```python
df['y'].value_counts().get('yes')
```

```python
Numbers=df.select_dtypes(include=['number'])
Corr_Matrix = round(Numbers.corr(),4)
Corr_Matrix
```

```python
correlation_matrix=Numbers.corr(method='pearson')


# generating heatmap
plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm',center=0)
plt.title("Correlation Matrix")
plt.show()
```
